```python
import pandas as pd
data = pd.read_csv('diabetes.csv')
data
```

```python
a = data.iloc[:,:8]
input_ = a.values.tolist()
input_
```

```python
output = data['Outcome'].tolist()
output
```

```python
import numpy as np
w = np.zeros((768,9))
w
```

```python
import random
```

```python
for i in range(768):
    for j in range(9):
        a = round(random.uniform(0, 1), 2)
        w[i][j] = a
w
```

```python
b = round(random.uniform(0, 1), 2)
b
```

```python
def step_function(x):
    if x > 0:
        return 1
    else:
        return 0
```

```python
def error(out,i):
    return output[i] - out
```

```python
errors = np.zeros((768,9)).astype(float)
errors
```

```python
y_pred = []
for i in range(768):
    sum = 0
    out = 0
    a = 0
    for j in range(8):
        sum += input_[i][j]*w[i][j] + b
    out = step_function(sum)
    a = error(out,i)
    if a != 0:
        errors[i][j] = a
    while a != 0:
        for j in range(8):
            sum += input_[i][j]*w[i][j] + b
        out = step_function(sum)
        a = error(out,i)
        if a != 0:
            errors[i][j] = a 
        for j in range(8):
            w[i][j] = w[i][j] + a*output[i]
        b = b + a
    y_pred.append(out)
    print("Diabetes({}) = {}".format(input_[i],out))
```
